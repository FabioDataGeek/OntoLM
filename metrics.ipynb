Creamos las funciones para ver F1, Accuracy y recall de las clasificación multilabel a partir de los archivos de JSON obtenidos

```python
import json
import pickle as pkl
```

```python
with open('/usrvol/resultados/categórico correcto 17-11-2023/train_dev_test_data.json', 'r') as f:
    epochs = json.load(f)
```

```python
epochs
```

```python
# GLOBAL ACCURACY
dev_dict = {}
test_dict = {}

for epoch in epochs:
    global_true_negatives = 0
    global_false_negatives = 0
    global_false_positives = 0
    global_true_positives = 0
    dev_matrices = epoch['dev_acc']
    for negatives in dev_matrices[0]:
        global_true_negatives += negatives[0]
        global_false_negatives += negatives[1]
    for positives in dev_matrices[1]:
        global_false_positives += positives[0]
        global_true_positives += positives[1]
    dev_dict[epoch['epoch']] = {'global_true_negatives': global_true_negatives,
                                'global_false_negatives': global_false_negatives,
                                'global_false_positives': global_false_positives,
                                'global_true_positives': global_true_positives}
    
    global_true_negatives = 0
    global_false_negatives = 0
    global_false_positives = 0
    global_true_positives = 0
    test_matrices = epoch['test_acc']
    for negatives in test_matrices[0]:
        global_true_negatives += negatives[0]
        global_false_negatives += negatives[1]
    for positives in test_matrices[1]:
        global_false_positives += positives[0]
        global_true_positives += positives[1]
    test_dict[epoch['epoch']] = {'global_true_negatives': global_true_negatives,
                                'global_false_negatives': global_false_negatives,
                                'global_false_positives': global_false_positives,
                                'global_true_positives': global_true_positives}

dev_accuracies = []
test_accuracies = []

for epoch in epochs:
    true_negative = dev_dict[epoch['epoch']]['global_true_negatives']
    false_negative = dev_dict[epoch['epoch']]['global_false_negatives']
    false_positive = dev_dict[epoch['epoch']]['global_false_positives']
    true_positive = dev_dict[epoch['epoch']]['global_true_positives']
    dev_accuracy = (true_negative + true_positive) / (true_negative + false_negative + false_positive + true_positive)
    dev_accuracies.append(dev_accuracy)

    true_negative = test_dict[epoch['epoch']]['global_true_negatives']
    false_negative = test_dict[epoch['epoch']]['global_false_negatives']
    false_positive = test_dict[epoch['epoch']]['global_false_positives']
    true_positive = test_dict[epoch['epoch']]['global_true_positives']
    test_accuracy = (true_negative + true_positive) / (true_negative + false_negative + false_positive + true_positive)
    test_accuracies.append(test_accuracy)
```

```python
# metrics by class

dev_per_epoch = {}
test_per_epoch = {}

for epoch in epochs:
    dev_true_negatives = []
    dev_false_negatives = []
    dev_false_positives = []
    dev_true_positives = []
    dev_matrices = epoch['dev_acc']
    for dev_matrix in dev_matrices:
        dev_true_negatives.append(dev_matrix[0][0])
        dev_false_negatives.append(dev_matrix[0][1])
        dev_false_positives.append(dev_matrix[1][0])
        dev_true_positives.append(dev_matrix[1][1])

    dev_per_epoch[epoch['epoch']] = {'true_negatives': dev_true_negatives,
                          'false_negatives': dev_false_negatives,
                          'false_positives': dev_false_positives,
                          'true_positives': dev_true_positives}

    test_true_negatives = []
    test_false_negatives = []
    test_false_positives = []
    test_true_positives = []
    test_matrices = epoch['test_acc']
    for test_matrix in test_matrices:
        test_true_negatives.append(test_matrix[0][0])
        test_false_negatives.append(test_matrix[0][1])
        test_false_positives.append(test_matrix[1][0])
        test_true_positives.append(test_matrix[1][1])
    
    test_per_epoch[epoch['epoch']] = {'true_negatives': test_true_negatives,
                            'false_negatives': test_false_negatives,
                            'false_positives': test_false_positives,
                            'true_positives': test_true_positives}
```

```python
acc_per_epoch = {}
prec_per_epoch = {}
rec_per_epoch = {}

for epoch in epochs:
    #accuracy by class
    dev_accuracies = []
    test_accuracies = []

    #recall by class
    dev_recalls = []
    test_recalls = []

    # precision by class
    dev_precisions = []
    test_precisions = []
    for i in range(40):
        true_negative = dev_per_epoch[epoch['epoch']]['true_negatives'][i]
        false_negative = dev_per_epoch[epoch['epoch']]['false_negatives'][i]
        false_positive = dev_per_epoch[epoch['epoch']]['false_positives'][i]
        true_positive = dev_per_epoch[epoch['epoch']]['true_positives'][i]
        
        dev_accuracy = (true_negative + true_positive) / (true_negative + false_negative + false_positive + true_positive)
        dev_accuracies.append(dev_accuracy)
        try:
            dev_recall = true_positive / (true_positive + false_negative)
        except ZeroDivisionError:
            dev_recall = 0
        dev_recalls.append(dev_recall)
        try:
            dev_precision = true_positive / (true_positive + false_positive)
        except ZeroDivisionError:
            dev_precision = 0
        dev_precisions.append(dev_precision)

        true_negative = test_per_epoch[epoch['epoch']]['true_negatives'][i]
        false_negative = test_per_epoch[epoch['epoch']]['false_negatives'][i]
        false_positive = test_per_epoch[epoch['epoch']]['false_positives'][i]
        true_positive = test_per_epoch[epoch['epoch']]['true_positives'][i]

        test_accuracy = (true_negative + true_positive) / (true_negative + false_negative + false_positive + true_positive)
        test_accuracies.append(test_accuracy)
        try:
            test_recall = true_positive / (true_positive + false_negative)
        except ZeroDivisionError:
                test_recall = 0
        test_recalls.append(test_recall)
        try:
            test_precision = true_positive / (true_positive + false_positive)
        except ZeroDivisionError:
            test_precision = 0
        test_precisions.append(test_precision)
    acc_per_epoch[epoch['epoch']] = {'dev': dev_accuracies, 'test': test_accuracies}
    prec_per_epoch[epoch['epoch']] = {'dev': dev_precisions, 'test': test_precisions}
    rec_per_epoch[epoch['epoch']] = {'dev': dev_recalls, 'test': test_recalls}
```
