# Análisis de relaciones

```python
import pickle as pkl
import networkx as nx
```

```python
with open('/usrvol/mappings/graph.pickle', 'rb') as f:
    G = pkl.load(f)
```

```python
relations_dict = {}

for n, nbrs in G.adjacency():
    for nbr, edict in nbrs.items():
        for key, eattr in edict.items():
            if eattr['rel'] not in relations_dict:
                relations_dict[eattr['rel']] = 1
            else:
                relations_dict[eattr['rel']] += 1
    
```

```python
relations_dict
```

```python
total_number = 0
for key, value in relations_dict.items():
    total_number += value
```

```python
total_number
```

```python
top_relations = sorted(relations_dict.items(), key=lambda x: x[1], reverse=True)[:10]
```

Aquí podemos ver el top 50 de relaciones, el cual constituye el 95% de las relaciones totales, esto quiere decir que podemos tomar estas manteniendo casi todo el grafo y nos libraríamos de emplear 1140 relaciones diferentes, tomando solo las más utilizadas. 

```python
top_relations
```

```python
total_number_filtered = sum([x[1] for x in top_relations])
```

```python
total_number_filtered
```

```python
total_number_filtered/total_number
```

Comprobamos cuales son las relaciones más utilizadas de las seleccionadas, ya que por cada relación que hemos seleccionado deberíamos incluir también su relación inversa

```python
with open('/usrvol/mappings/IdentifierToRelationsTotal.pickle', 'rb') as f:
    IdentifierToRelationsTotal = pkl.load(f)
```

```python
for relation in top_relations:
    print(IdentifierToRelationsTotal[relation[0]])
```
