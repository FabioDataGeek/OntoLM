# An√°lisis de relaciones

```python
import pickle as pkl
import networkx as nx
```

```python
with open('/usrvol/mappings/graph.pickle', 'rb') as f:
    G = pkl.load(f)
```

```python
relations_dict = {}

for n, nbrs in G.adjacency():
    for nbr, edict in nbrs.items():
        for key, eattr in edict.items():
            if eattr['rel'] not in relations_dict:
                relations_dict[eattr['rel']] = 1
            else:
                relations_dict[eattr['rel']] += 1
    
```

```python
relations_dict
```

```python
total_number = 0
for key, value in relations_dict.items():
    total_number += value
```

```python
total_number
```

```python
top_relations = sorted(relations_dict.items(), key=lambda x: x[1], reverse=True)[:10]
```

Aqu√≠ podemos ver el top 50 de relaciones, el cual constituye el 95% de las relaciones totales, esto quiere decir que podemos tomar estas manteniendo casi todo el grafo y nos librar√≠amos de emplear 1140 relaciones diferentes, tomando solo las m√°s utilizadas. 

```python
top_relations
```

```python
total_number_filtered = sum([x[1] for x in top_relations])
```
