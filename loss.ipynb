```python
import pickle as pkl
import json
import matplotlib.pyplot as plt
```

```python
with open('/usrvol/resultados/categ√≥rico correcto 17-11-2023/loss_lr.json', 'r') as f:
    data = json.load(f)
```

```python
iterator = None
data_per_epoch = []
for i in range(len(data)):
    epoch = data[i]['epoch']
    if iterator != epoch:
        iterator = epoch
        data_per_epoch.append([])
    data_per_epoch[-1].append(data[i]['loss'])
```
